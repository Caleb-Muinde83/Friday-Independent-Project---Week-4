# 1.	Identify the most popular hour of the day
# for picking up a shared electric car (Bluecar)
# in the city of Paris over the month of April 2018.

```python
# Select Rows based on Paris in a city
subsetDataFrame = df[df['City'] == 'Paris']
print("DataFrame with City : paris" , subsetDataFrame.to_string(), sep='\n')
```

```python
# •	Dropping the Displayed comment  and  Scheduled at columns due to lots of NaN values.

subsetDataFrame.drop(['Scheduled at','Displayed comment'], axis=1)
print("DataFrame with City : paris" , subsetDataFrame.to_string(), sep='\n')
```

```python
# Find the popular hour of the hour for picking up a shared electric car (Bluecar) in the city of Paris
deptCounts = subsetDataFrame.hour.value_counts()
print("-------Frequency counts for hour column-------\n",deptCounts)
print(type(deptCounts))
```
